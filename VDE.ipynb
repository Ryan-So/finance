## Extracting the Data

Pandas has a delightful api for accessing stock data via Yahoo or Google. There are, however, the following gotcha:

1. The start date defaults to 1/1/2010 so make sure that you pick a proper starting date. Note that you can pick a silly early date (e.g., `1800/1/1`) to get everything.
2. The file automatically extracts, start, close, low, high, and volume information so if you are only interested in one, be sure to extract that straight off. 

```python
import pandas.io.data as web
import Quandl
```

```python
VDE = web.DataReader("VDE", 'yahoo', start='1800/1/1')['Close']
SPY = web.DataReader('SPY', 'yahoo', start='1800/1/1')['Close']
XOM = web.DataReader('XOM', 'yahoo', start='1800/1/1')['Close']
```

```python
BRENT = Quandl.get("OFDP/FUTURE_B1", trim_start="1990-08-17", trim_end="2014-11-24")['Settle']
```
